```python
dfs = [df14,df15,df16,df17,df18,df19,df20]
subgenres = ['Introspective', 'Club (Materialistic)','Romance','Mumble ', 'Old School Gangsta Rap', 'New School Gangsta Rap']
percent_intro = []
percent_club = []
percent_romance = []
percent_mumble = []
percent_oldschool = []
percent_newschool = []
for df in dfs:
    for sub in subgenres:
        if sub == 'Introspective':
            percent_intro.append(len(df[df['subgenre']==sub])/len(df))
        elif sub == 'Club (Materialistic)':
            percent_club.append(len(df[df['subgenre']==sub])/len(df))
        elif sub == 'Romance':
            percent_romance.append(len(df[df['subgenre']==sub])/len(df))
        elif sub == 'Mumble ':
            percent_mumble.append(len(df[df['subgenre']==sub])/len(df))
        elif sub == 'Old School Gangsta Rap':
            percent_oldschool.append(len(df[df['subgenre']==sub])/len(df))
        else:
            percent_newschool.append(len(df[df['subgenre']==sub])/len(df))
```

```python
years = [2014,2015,2016,2017,2018,2019,2020]
```

```python
fig,ax = plt.subplots(figsize=(20,6))

ax.plot(years,percent_intro,label="Lyrical")
ax.plot(years,percent_club,label="Club")
ax.plot(years,percent_romance,label="Romance")
ax.plot(years,percent_mumble,label = "Mumble Rap")
ax.plot(years,percent_oldschool,label = "Old School Gangsta Rap")
ax.plot(years,percent_newschool, label = "New School Street Rap")
ax.set_ylabel('% of top 10 rap songs',size=15)
ax.legend()
plt.savefig('images/over_time.png')
```
